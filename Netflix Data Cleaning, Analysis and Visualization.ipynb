# Comparison between movies and TV shows

```python
grp1=df.groupby(by='Type')
def length(a):
   return len(a) 
series1=grp1.apply(length)
plt.style.use('fast')
xpara=series1.index
ypara=series1.values
plt.pie(x=ypara,labels=xpara,autopct='%1.2f%%',pctdistance=0.7)
plt.pie(x=[1],radius=0.5,colors='w')
plt.legend()
plt.show()
```

Movies are released more in number than tv shows.

# Release count for Movies and TV Shows through the year

```python
grp2=df.groupby(by=['Release_year','Type'])
##grp3=df.groupby(by=['Release_year'])
ser2=grp2.apply(len)
##ser3=grp3.apply(len)
serr2=ser2.unstack(level=1)
serr2['Movie']=serr2['Movie'].fillna(0).astype(int)
serr2['TV Show']=serr2['TV Show'].fillna(0).astype(int)
a=serr2['Movie']
b=serr2['TV Show']
xpar=serr2.index
y1=a.values
y2=b.values
plt.figure(figsize=(10,15))
width=0.3
plt.plot(xpar,y1,label='Movie')
plt.plot(xpar,y2,label='TV Show',color='y')
##plt.xticks(np.arange(1920,10,2020))
plt.xlabel('Year_Released')
plt.ylabel('Count')
plt.xticks(rotation=80)
plt.xticks(ticks=np.arange(1920,2025,3))
plt.legend()
plt.show()
```

Movies has always been dominant 

# Country wise Movie and TV Shows visualisation.Which country has produced how many movies and tv shows?

```python
xpara=df['Country'].unique()
grp3=df.groupby(['Country','Type'])
gr3=grp3.apply(len).unstack(level=1).fillna(0).astype(int)
c=gr3['Movie']
d=gr3['TV Show']
xpar=gr3.index
y1=list(c.values)
y2=list(d.values)
y3=[y1[i]+y2[i] for i in range(len(y1))]
plt.figure(figsize=(20,10))
plt.bar(x=xpar,height=y1,label='Movie')
plt.bar(x=xpar,height=y2,bottom=y1,label='TV Show')
plt.xlabel('Country')
plt.ylabel('Type count')
plt.legend(loc=0)
plt.xticks(rotation=90)
plt.title('Country and its movie vs tvshows ratio')
plt.show()
```

United States tops the chart in both movie count and tv shows count

# Which month has the most releases added?

```python
##Adding date year and month columns from the dataset
df['Year']=df['Date_added'].dt.year
df['Month']=df['Date_added'].dt.month
df['Day']=df['Date_added'].dt.day
df
```

```python
gr4=df.groupby(by='Month')
grp4=gr4.apply(len)
grp4
plt.stem(grp4.index,grp4.values,linefmt='--',basefmt='C2-')
plt.xticks(ticks=np.arange(1,13,1),labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug',
 'Sep', 'Oct', 'Nov', 'Dec'])
plt.yticks(ticks=np.arange(0,1000,100))
for x, y in zip(grp4.index, grp4.values):
 # the position of the data label relative to the data point can be adjusted by adding/subtracting a value from the x &/ y coordinates
  plt.text(x = x,y = y+30,s='{:.0f}'.format(y),color ='black')
plt.xlabel('Months')
plt.ylabel('Added content counts')
plt.title('Months vs release added')
plt.show()
```

July month has more contents added to Netflix.

# Release of Movies vs TV Shows across months.

```python
moviedf=df[df['Type']=='Movie']
movdf=moviedf.groupby(by='Month')
movplot=movdf.apply(len)
tvsdf=df[df['Type']=='TV Show']
tvdf=tvsdf.groupby(by='Month')
tvplot=tvdf.apply(len)
x=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug',
 'Sep', 'Oct', 'Nov', 'Dec']
plt.plot(x,movplot.values,label='Movie')
plt.plot(x,tvplot.values,label='TVShow')
plt.grid()
plt.xlabel('Months')
plt.ylabel('Released counts')
plt.title('Release of movievs tvshows across months')
plt.legend()
plt.show()
```

Movie releases tops all the months.

# Ratings vs no of contents rated as them.

```python
ratinggrp=df.groupby(by='Rating')
rating=ratinggrp.apply(len).sort_values(ascending=False)
rating
ypara=rating.values
xpara=rating.index
Ratingplot=sns.stripplot(data=rating,x=xpara,y=rating.values)
for x, y in zip(xpara, ypara):
 # the position of the data label relative to the data point can be adjusted by adding/subtracting a value from the x &/ y coordinates
  plt.text(x = x,y = y+50,s='{:.0f}'.format(y),color ='purple') # set colour of line
Ratingplot.set_title('Rating and its count')
Ratingplot.set_xlabel('Rating')
Ratingplot.set_ylabel('count')
Ratingplot.set_xticks(ticks=xpara,labels=xpara,rotation=90)
```

TV_MA has the highest no of contents rated as it.

# Year wise content release.

```python
plt.figure(figsize=(10,10))
yeargrp=df.groupby(by='Release_year')
yrgrp=yeargrp.apply(len).sort_values(ascending=False).head(20).astype(int)
plt.plot(yrgrp.index,yrgrp.values,marker='o')
plt.title('No of Content released each year')
plt.xlabel('Year of Release')
plt.ylabel('Content count')
for x, y in zip(yrgrp.index, yrgrp.values):
 # the position of the data label relative to the data point can be adjusted by adding/subtracting a value from the x &/ y coordinates
  plt.text(x = x,y = y+20,s='{:.0f}'.format(y),color ='purple')
plt.xticks(ticks=np.arange(2002,2022,1),rotation=85)
yrgrp
```

From the above visualisation it is evident that contents began to get released drastically after 2015 
and began dropping after 2019.
